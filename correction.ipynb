```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt('data/ECGu.txt') # Load the ECG data from a text file
print(data)
print(data.shape) # (number of rows, number of columns)
```

```python
# Conversion of data from A/U to mV (1 A/U = 1.024 mV)
data = data * 1.024
```

```python
# create a time vector t at the same lenghth as EGC and with a sampling rate of 1000 Hz
fs = 1000 # sampling rate in Hz
t = np.arange(0, data.shape[0])/fs
print(t)
```

```python
# Plot the ECG data as a function of time
plt.plot(t, data)
plt.xlabel('Time (s)')
```
