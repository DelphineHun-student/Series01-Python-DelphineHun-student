```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
```

## Load the data

```python
data = np.loadtxt('data/ECGu.txt') # Load the ECG data from a text file
print(data)
print(data.shape) # (number of rows, number of columns)
```

## Time definition

```python
ecg_mv = data * 1.024/1000 # Convert to mV

fs = 1000 # sampling rate in Hz
t = np.arange(0, data.shape[0])/fs
```

# Realistic ECG plot

```python
# Charger les données (suppose 4 colonnes : temps + 3 amplitudes)
data = np.loadtxt('data/ECGu.txt')  
fig, axs = plt.subplots(3, 1, figsize=(10, 8), sharex=True) # Create 3 subplots sharing x-axis

time = t     # Time
amplitude1 = ecg_mv[:, 0]   # Colonne 1 (index 0)
amplitude2 = ecg_mv[:, 1]   # Colonne 2 (index 1)
amplitude3 = ecg_mv[:, 2]  # Colonne 3 (index 2)

axs[0].plot(time, amplitude1)
axs[1].plot(time, amplitude2)
axs[2].plot(time, amplitude3)

axs[0].set_ylabel('I', rotation=0)
axs[1].set_ylabel('II', rotation=0)
axs[2].set_ylabel('III', rotation=0)
axs[2].set_xlabel('Temps (s)')

for ax in axs: # ax is a sous graphique
    ax.grid(which='major', color='red', linestyle='-', linewidth=0.5)
    ax.grid(which='minor', color='pink', linestyle='-', linewidth=0.5)
    ax.minorticks_on()

plt.figure(figsize=(10, 5))

plt.tight_layout
plt.show()
```
