# ECG plot

```python
# Charger les données (suppose 4 colonnes : temps + 3 amplitudes)
data = np.loadtxt('data/ECGu.txt', skiprows=1)  
fig, axs = plt.subplots(3, 1, figsize=(10, 8), sharex=True) # Create 3 subplots sharing x-axis

time = data[:, 0]        # Colonne 1 (index 0)
amplitude1 = data[:, 3]   # Colonne 4 (index 3)
amplitude2 = data[:, 4]   # Colonne 5 (index 4)
amplitude3 = data[:, 5]  # Colonne 6 (index 5)

axs[0].plot(time, amplitude1)
axs[1].plot(time, amplitude2)
axs[2].plot(time, amplitude3)

axs[0].set_ylabel('I', rotation=0)
axs[1].set_ylabel('II', rotation=0)
axs[2].set_ylabel('III', rotation=0)
axs[2].set_xlabel('Temps (s)')

for ax in axs: # ax is a sous graphique
    ax.grid(which='major', color='red', linestyle='-', linewidth=0.5)
    ax.grid(which='minor', color='pink', linestyle='-', linewidth=0.5)
    ax.minorticks_on()

plt.figure(figsize=(10, 5))

plt.tight_layout
plt.show()
```
